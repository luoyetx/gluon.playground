```python
%matplotlib inline
import time
import cv2
import numpy as np
import mxnet as mx
from mxnet import nd, autograd as ag, gluon as gl
from mxnet.gluon import nn
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
mx.random.seed(0)


class Generator(gl.HybridBlock):
    
    def __init__(self, num_filters=64):
        super(Generator, self).__init__()
        self.num_filters = num_filters
        with self.name_scope():
            self.prev = nn.HybridSequential()
            self.prev.add(nn.Dense(1024, use_bias=False), nn.BatchNorm(), nn.Activation(activation='relu'))
            self.G = nn.HybridSequential()
            self.G.add(self.Upsample2D(2*num_filters, 7, 1, 0))  # 7x7
            self.G.add(self.Upsample2D(num_filters, 4, 2, 1))   # 14x14
            self.G.add(self.Upsample2D(1, 4, 2, 1, activation='tanh'))   # 28x28
    
    def Upsample2D(self, num_filters, kernel, stride, pad, activation='relu'):
        block = nn.HybridSequential()
        block.add(nn.Conv2DTranspose(num_filters, kernel, stride, pad, use_bias=False))
        block.add(nn.BatchNorm())
        block.add(nn.Activation(activation=activation))
        return block
    
    def hybrid_forward(self, F, x):
        x = self.prev(x)
        x = F.reshape(x, (0, -1, 1, 1))
        return self.G(x)


class Discriminator(gl.HybridBlock):
    
    def __init__(self, num_filters=64, num_category=10, num_continuous=2):
        super(Discriminator, self).__init__()
        with self.name_scope():
            self.D = nn.HybridSequential()
            self.D.add(self.DownSample2D(num_filters))   # 14x14
            self.D.add(self.DownSample2D(2*num_filters)) # 7x7
            self.D.add(nn.Dense(1024, use_bias=False), nn.BatchNorm(), nn.Activation(activation='relu'))
            self.prob = nn.Dense(1)
            # info
            self.feat = nn.HybridSequential()
            self.feat.add(nn.Dense(128, use_bias=False), nn.BatchNorm(), nn.Activation(activation='relu'))
            self.cate_prob = nn.Dense(num_category)
            self.cont_mean = nn.Dense(num_continuous)
            self.Q = nn.HybridSequential()
            self.Q.add(self.feat, self.cate_prob, self.cont_mean)
    
    def DownSample2D(self, num_filters):
        block = nn.HybridSequential()
        block.add(nn.Conv2D(num_filters, kernel_size=4, strides=2, padding=1, use_bias=False))
        block.add(nn.BatchNorm())
        block.add(nn.Activation(activation='relu'))
        return block
            
    def hybrid_forward(self, F, x):
        x = self.D(x)
        prob = self.prob(x)
        feat = self.feat(x)
        cate_prob = self.cate_prob(feat)
        cont_mean = self.cont_mean(feat)
        return prob, cate_prob, cont_mean
    
    def Q_params(self):
        return self.Q.collect_params()


class GaussianNNL(gl.loss.Loss):
    
    def __init__(self, eps=1e-8):
        super(GaussianNNL, self).__init__(None, 0)
        self.eps = 1e-8
        
    def hybrid_forward(self, F, x, mean, var):
        var = var + self.eps
        loss = F.log(2*np.pi*var) + F.square(x - mean) / var
        loss = loss / 2
        return F.mean(loss, axis=self._batch_axis, exclude=True)

    
batch_size = 32
epoches = 30
ctx = mx.gpu()
z_dim = 62
num_category = 10
num_continuous = 2
num_filters = 16
lr1 = 1e-3
lr2 = 1e-4
beta1 = 0.5


def transform(data, label):
    data = nd.array(data.reshape((1, 28, 28))).astype(np.float32) / 128 - 1
    return data, label

train_data = gl.data.DataLoader(gl.data.vision.MNIST(root='./data', train=True, transform=transform),
                                batch_size=batch_size, shuffle=True, last_batch='discard')

netD = Discriminator(num_filters)
netD.hybridize()
netD.initialize(mx.init.Normal(0.01), ctx=ctx)
trainerD = gl.Trainer(netD.collect_params(), 'adam', {'learning_rate': lr2, 'beta1': beta1})
netG = Generator(num_filters)
netG.hybridize()
netG.initialize(mx.init.Normal(0.01), ctx=ctx)
trainerG = gl.Trainer(netG.collect_params(), 'adam', {'learning_rate': lr1, 'beta1': beta1})
trainerQ = gl.Trainer(netD.Q_params(), 'adam', {'learning_rate': lr1, 'beta1': beta1})

criterion1 = gl.loss.SigmoidBinaryCrossEntropyLoss()
criterion2 = gl.loss.SoftmaxCrossEntropyLoss()
criterion3 = gl.loss.L2Loss()

real_label = nd.ones(batch_size, ctx=ctx)
fake_label = nd.zeros(batch_size, ctx=ctx)


def sampleG(netG, num=64):
    # sample data
    z = nd.random.normal(0, 1, shape=(num, z_dim)).as_in_context(ctx)
    label = nd.array(np.random.randint(num_category, size=num)).as_in_context(ctx)
    c1 = nd.one_hot(label, depth=num_category).as_in_context(ctx)
    c2 = nd.random.uniform(-1, 1, shape=(num, num_continuous)).as_in_context(ctx)
    z = nd.concat(z, c1, c2, dim=1)
    x = netG(z)
    x = (x + 1) * 128
    x = x.asnumpy()
    return x


def sampleG_usingC(netG):
    # sample data
    z = np.random.normal(0, 1, size=(10, z_dim))
    z = nd.array(np.tile(z, (1, 10)).reshape(100, z_dim))
    label = [[i]*10 for i in range(10)]
    label = nd.array(np.array(label).reshape(-1))
    c1 = nd.one_hot(label, depth=num_category)
    c2 = np.linspace(-1, 1, 10)
    c2 = np.vstack([c2, c2]).T
    c2 = nd.array(np.tile(c2, (10, 1)))
    for i in range(5):
        c2[10*i:10*(i+1), 0] = np.random.uniform(-1, 1)
    for i in range(5, 10):
        c2[10*i:10*(i+1), 1] = np.random.uniform(-1, 1)
    z = nd.concat(z, c1, c2, dim=1).as_in_context(ctx)
    x = netG(z)
    x = (x + 1) * 128
    x = x.asnumpy()
    return x


def visImages(data):
    data = np.clip(data, 0, 255).astype(np.uint8)
    n = int(np.ceil(np.sqrt(data.shape[0])))
    h, w = data.shape[2:]
    canvas = np.zeros((n*h, n*w), dtype=np.uint8)
    for i in range(data.shape[0]):
        y, x = i / n, i % n
        canvas[y*h:y*h+h, x*w:x*w+w] = data[i, 0]
    plt.figure(figsize=(10, 10))
    plt.imshow(canvas, cmap='gray')
    plt.show()


for epoch in range(epoches):
    etic = time.time()
    moving_lossD = 0
    moving_lossG = 0
    moving_lossGc1 = 0
    moving_lossGc2 = 0
    for i, (data, label) in enumerate(train_data):
        real_x = data.as_in_context(ctx)
        # sample data
        z = nd.random.normal(0, 1, shape=(batch_size, z_dim)).as_in_context(ctx)
        label = nd.array(np.random.randint(num_category, size=batch_size)).as_in_context(ctx)
        c1 = nd.one_hot(label, depth=num_category).as_in_context(ctx)
        c2 = nd.random.uniform(-1, 1, shape=(batch_size, num_continuous)).as_in_context(ctx)
        z = nd.concat(z, c1, c2, dim=1)
        # train netD
        with ag.record():
            fake_x = netG(z)
            fake_y, _, _ = netD(fake_x.detach())
            real_y, _, _ = netD(real_x)
            lossD = criterion1(fake_y, fake_label) + criterion1(real_y, real_label)
        lossD.backward()
        trainerD.step(batch_size)
        # train netG
        with ag.record():
            fake_y, cate_prob, cont_mean = netD(fake_x)
            lossG = criterion1(fake_y, real_label)
            lossGc1 = criterion2(cate_prob, label)
            lossGc2 = criterion3(c2, cont_mean)
            lossGAll = lossG + lossGc1 + lossGc2
        lossGAll.backward()
        trainerG.step(batch_size)
        trainerQ.step(batch_size)
        # log
        moving_lossD += lossD.mean().asscalar()
        moving_lossG += lossG.mean().asscalar()
        moving_lossGc1 += lossGc1.mean().asscalar()
        moving_lossGc2 += lossGc2.mean().asscalar()
        if i % 100 == 99:
            moving_lossD /= 100
            moving_lossG /= 100
            moving_lossGc1 /= 100
            moving_lossGc2 /= 100
            print('[epoch %03d][iter %03d] lossD: %f, lossG: %f, lossGc1: %f, lossGc2: %f' %
                  (epoch+1, i+1, moving_lossD, moving_lossG, moving_lossGc1, moving_lossGc2))
            moving_lossD = 0
            moving_lossG = 0
            moving_lossGc1 = 0
            moving_lossGc2 = 0
    etoc = time.time()
    print('[epoch %03d] costs %fs' % (epoch+1, etoc-etic))
    visImages(sampleG_usingC(netG))
    netG.save_params('./tmp/mnist-netG-%04d.params' % (epoch+1))
```

```python
visImages(sampleG_usingC(netG))
visImages(sampleG(netG, 100))
```
